```python
size=(25, 25)
data=np.zeros((len(x[:,0]),25,25))
nn=0

for i in (x[:,0]):
    print str(int(i))+".jpg"
    file=str(int(i))+".jpg"
    jpgfile = Image. open(file)
    r,g,b = jpgfile.split()
    p_resized = r.resize(size, Image.ANTIALIAS)
    pp=np.reshape(list(p_resized.getdata()),p_resized.size)
    data[nn,:,:]=pp/255.
    nn=nn+1
```

```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
keras.backend.clear_session()
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(25, 25)),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dense(3, activation=tf.nn.softmax)])
model.summary()

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
train_data=data[:50000,:,:]
test_data=data[50000:,:,:]
train_label=type[:50000]
test_label=type[50000:]

model.fit(train_data, train_label, epochs=6)
```

```python
test_loss, test_acc = model.evaluate(test_data, test_label)
print('Test accuracy:', test_acc)
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
        rotation_range=180,
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest')

#test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(25, 25, 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(3))
model.add(Activation('relu'))

model.summary()

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```
